```python
import numpy as np
import matplotlib.pyplot as plt

# 假设有两组数据和它们的标准差
categories = ['0.1', '0.01', '0.001']
bar_data1 = [84.29, 85.25, 85.77]
bar_std1 = [1.80, 1.47, 1.31]

bar_data2 = [83.54, 85.21, 85.67]
bar_std2 = [1.74, 1.26, 1.79]

# 创建第一个图形和左边y轴绘制柱状图的轴对象
# fig1, ax1 = plt.subplots()
fig1, ax1 = plt.subplots(figsize=(5, 8), dpi=300) 
bar_width = 0.4
bar_positions1 = np.arange(len(categories))
bar_positions2 = [pos + bar_width for pos in bar_positions1]
plt.xlabel('$\epsilon$', fontsize=21)
ax1.bar(bar_positions1, bar_data1, yerr=bar_std1, capsize=4, width=bar_width, color='g', label='$R^2LP$', align='center')

ax1.bar(bar_positions2, bar_data2, yerr=bar_std2, capsize=4, width=bar_width, color='r', label='$Opt-LP$', align='center')

ax1.set_xticks([pos + bar_width / 2 for pos in bar_positions1])
ax1.set_xticklabels(categories)
# ax1.set_ylabel('Bar Chart with Error Bars')
# ax1.set_ylim(20, max(max(bar_data1), max(bar_data2)) + max(max(bar_std1), max(bar_std2)) + 5)
ax1.legend(loc='best', fontsize=21)
ax1.set_ylabel('Accuracy', fontsize=21)
ax1.tick_params(axis='y', labelsize=21)
ax1.tick_params(axis='x', labelsize=21)
ax1.set_aspect('auto', adjustable='box')
# 显示图形
ax1.set_ylim(70, 90) 

# 创建第二个图形和右边y轴绘制柱状图的轴对象
# fig2, ax2 = plt.subplots()
fig2, ax2 = plt.subplots(figsize=(5, 8), dpi=300) 
bar_width = 0.4
line_data1 = [0.019866, 0.152076, 0.8899]
# line_std1 = bar_std1

line_data2 = [0.0499, 0.530, 4.612]
ax2.bar(bar_positions1, line_data1, width=bar_width, color='r', label='$R^2LP$', align='center')
ax2.bar(bar_positions2, line_data2, width=bar_width, color='m', label='$Opt-LP$', align='center')
plt.xlabel('$\epsilon$', fontsize=21)
ax2.set_xticks([pos + bar_width / 2 for pos in bar_positions1])
ax2.set_xticklabels(categories)
ax2.set_ylabel('Query Time (s)', fontsize=21)
# ax2.set_ylim(20, max(max(bar_data1), max(bar_data2)) + 5)
ax2.legend(loc='best', fontsize=21)
ax2.tick_params(axis='y', labelsize=21)
ax2.tick_params(axis='x', labelsize=21)
ax2.set_aspect('auto', adjustable='box')
# 显示图形
fig1.savefig('cora-accu.png')  # Save the first figure as 'figure1.png'
fig2.savefig('cora-time.png')  # Save the second figure as 'figure2.png'

plt.show()
```
